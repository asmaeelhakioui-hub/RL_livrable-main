## A small recap of Q-Learning

*Q-Learning* **is the RL algorithm that**:

- Trains *Q-Function*, an **action-value function** that encoded, in internal memory, by a *Q-table* **that contains all the state-action pair values.**

- Given a state and action, our Q-Function **will search the Q-table for the corresponding value.**
    
<img src="https://huggingface.co/datasets/huggingface-deep-rl-course/course-images/resolve/main/en/unit3/Q-function-2.jpg" alt="Q function"  width="100%"/>

- When the training is done,**we have an optimal Q-Function, so an optimal Q-Table.**
    
- And if we **have an optimal Q-function**, we
have an optimal policy, since we **know for, each state, the best action to take.**

<img src="https://huggingface.co/datasets/huggingface-deep-rl-course/course-images/resolve/main/en/unit3/link-value-policy.jpg" alt="Link value policy"  width="100%"/>


But, in the beginning,¬†our **Q-Table is useless since it gives arbitrary value for each state-action pair¬†(most of the time we initialize the Q-Table to 0 values)**. But, as we‚Äôll¬†explore the environment and update our Q-Table it will give us better and better approximations

##Define the epsilon-greedy policy ü§ñ

Epsilon-greedy is the training policy that handles the exploration/exploitation trade-off.

The idea with epsilon-greedy:

- With *probability 1‚Ää-‚Ää…õ* : **we do exploitation** (i.e. our agent selects the action with the highest state-action pair value).

- With *probability …õ*: we do **exploration** (trying a random action).

As the training continues, we progressively **reduce the epsilon value since we will need less and less exploration and more exploitation.**

## Define the hyperparameters ‚öôÔ∏è

The exploration related hyperparamters are some of the most important ones.

- We need to make sure that our agent **explores enough of the state space** to learn a good value approximation. To do that, we need to have progressive decay of the epsilon.
- If you decrease epsilon too fast (too high decay_rate), **you take the risk that your agent will be stuck**, since your agent didn't explore enough of the state space and hence can't solve the problem.

````python
# Param√®tres d'entra√Ænement
n_training_episodes = 10000  # Nombre total d'√©pisodes
learning_rate = 0.7          # Taux d'apprentissage

# Param√®tres d'√©valuation
n_eval_episodes = 100        # Nombre d'√©pisodes de test

# Param√®tres de l'environnement
env_id = "FrozenLake-v1"     # Nom de l'environnement
max_steps = 99               # Max d'√©tapes par √©pisode
gamma = 0.95                 # Facteur d'actualisation (discount rate)
eval_seed = []               # Graine pour l'√©valuation

# Param√®tres d'exploration
max_epsilon = 1.0             # Probabilit√© d'exploration au d√©but
min_epsilon = 0.05            # Probabilit√© d'exploration minimale
decay_rate = 0.0005            # Taux de d√©croissance exponentielle pour epsilon
````

```
For episode in the total of training episodes:

Reduce epsilon (since we need less and less exploration)
Reset the environment

  For step in max timesteps:    
    Choose the action At using epsilon greedy policy
    Take the action (a) and observe the outcome state(s') and reward (r)
    Update the Q-value Q(s,a) using Bellman equation Q(s,a) + lr [R(s,a) + gamma * max Q(s',a') - Q(s,a)]

```

````python
# Fonction d'entra√Ænement de l'agent Q-Learning
def train(n_training_episodes, min_epsilon, max_epsilon, decay_rate, env, max_steps, Qtable):
  for episode in tqdm(range(n_training_episodes)):
    # Calculer la valeur d√©croissante d'epsilon
    epsilon = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*episode)
    # R√©initialiser l'environnement au d√©but de l'√©pisode
    state, info = env.reset()
    step = 0
    terminated = False
    truncated = False

    # Boucle pour chaque √©tape de l'√©pisode
    for step in range(max_steps):
      # Choisir l'action selon la politique epsilon-gloutonne
      action = epsilon_greedy_policy(Qtable, state, epsilon)

      # Ex√©cuter l'action et observer le nouvel √©tat et la r√©compense
      new_state, reward, terminated, truncated, info = env.step(action)

      # Mettre √† jour la Q-table selon l'√©quation de Bellman
      # Q(s,a) = Q(s,a) + lr * [R + gamma * max Q(s',a') - Q(s,a)]
      Qtable[state][action] = Qtable[state][action] + learning_rate * (reward + gamma * np.max(Qtable[new_state]) - Qtable[state][action])

      # Fin de l'√©pisode si termin√© ou tronqu√©
      if terminated or truncated:
        break

      # Mise √† jour de l'√©tat
      state = new_state
  return Qtable
````

## The evaluation method üìù

- We defined the evaluation method that we're going to use to test our Q-Learning agent.

````python
# Fonction d'√©valuation de l'agent
def evaluate_agent(env, max_steps, n_eval_episodes, Q, seed):
  """
  √âvalue l'agent pour ``n_eval_episodes`` √©pisodes et retourne la r√©compense moyenne et l'√©cart-type.
  """
  episode_rewards = []
  for episode in tqdm(range(n_eval_episodes)):
    if seed:
      state, info = env.reset(seed=seed[episode])
    else:
      state, info = env.reset()
    step = 0
    truncated = False
    terminated = False
    total_rewards_ep = 0

    for step in range(max_steps):
      # Choisir l'action optimale (gloutonne)
      action = greedy_policy(Q, state)
      new_state, reward, terminated, truncated, info = env.step(action)
      total_rewards_ep += reward

      if terminated or truncated:
        break
      state = new_state
    episode_rewards.append(total_rewards_ep)
  mean_reward = np.mean(episode_rewards)
  std_reward = np.std(episode_rewards)

  return mean_reward, std_reward
````

## Evaluate our Q-Learning agent üìà

````python
# Fonction pour enregistrer une vid√©o de l'agent
def record_video(env, Qtable, out_directory, fps=1):
  """
  G√©n√®re une vid√©o de replay de l'agent
  """
  images = []
  terminated = False
  truncated = False
  state, info = env.reset(seed=random.randint(0,500))
  img = env.render()
  images.append(img)
  while not terminated or truncated:
    # Prendre l'action optimale
    action = np.argmax(Qtable[state][:])
    state, reward, terminated, truncated, info = env.step(action)
    img = env.render()
    images.append(img)
  imageio.mimsave(out_directory, [np.array(img) for i, img in enumerate(images)], fps=fps)
````

````python
# Fonction pour envoyer le mod√®le sur le Hub
def push_to_hub(
    repo_id, model, env, video_fps=1, local_repo_path="hub"
):
    """
    √âvalue, g√©n√®re une vid√©o et t√©l√©verse le mod√®le sur le Hugging Face Hub.
    Cette m√©thode effectue le pipeline complet :
    - √âvaluation du mod√®le
    - G√©n√©ration de la fiche mod√®le (model card)
    - Enregistrement de la vid√©o replay
    - Envoi de tout sur le Hub
    """
    _, repo_name = repo_id.split("/")

    eval_env = env
    api = HfApi()

    # √âtape 1 : Cr√©er le d√©p√¥t
    repo_url = api.create_repo(
        repo_id=repo_id,
        exist_ok=True,
    )

    # √âtape 2 : T√©l√©charger les fichiers
    repo_local_path = Path(snapshot_download(repo_id=repo_id))

    # √âtape 3 : Sauvegarder le mod√®le
    if env.spec.kwargs.get("map_name"):
        model["map_name"] = env.spec.kwargs.get("map_name")
        if env.spec.kwargs.get("is_slippery", "") == False:
            model["slippery"] = False

    # S√©rialiser le mod√®le (Pickle)
    with open((repo_local_path) / "q-learning.pkl", "wb") as f:
        pickle.dump(model, f)

    # √âtape 4 : √âvaluer le mod√®le et construire le JSON des m√©triques
    mean_reward, std_reward = evaluate_agent(
        eval_env, model["max_steps"], model["n_eval_episodes"], model["qtable"], model["eval_seed"]
    )

    evaluate_data = {
        "env_id": model["env_id"],
        "mean_reward": mean_reward,
        "n_eval_episodes": model["n_eval_episodes"],
        "eval_datetime": datetime.datetime.now().isoformat()
    }

    # √âcrire le fichier results.json
    with open(repo_local_path / "results.json", "w") as outfile:
        json.dump(evaluate_data, outfile)

    # √âtape 5 : Cr√©er la model card
    env_name = model["env_id"]
    if env.spec.kwargs.get("map_name"):
        env_name += "-" + env.spec.kwargs.get("map_name")

    if env.spec.kwargs.get("is_slippery", "") == False:
        env_name += "-" + "no_slippery"

    metadata = {}
    metadata["tags"] = [env_name, "q-learning", "reinforcement-learning", "custom-implementation"]

    # Ajouter les m√©triques
    eval = metadata_eval_result(
        model_pretty_name=repo_name,
        task_pretty_name="reinforcement-learning",
        task_id="reinforcement-learning",
        metrics_pretty_name="mean_reward",
        metrics_id="mean_reward",
        metrics_value=f"{mean_reward:.2f} +/- {std_reward:.2f}",
        dataset_pretty_name=env_name,
        dataset_id=env_name,
    )

    # Fusionner les dictionnaires
    metadata = {**metadata, **eval}

    model_card = f"""
  # **Q-Learning** Agent playing1 **{env_id}**
  This is a trained model of a **Q-Learning** agent playing **{env_id}** .

  ## Usage

  ```python

  model = load_from_hub(repo_id="{repo_id}", filename="q-learning.pkl")

  # Don't forget to check if you need to add additional attributes (is_slippery=False etc)
  env = gym.make(model["env_id"])
  ```
  """

    evaluate_agent(env, model["max_steps"], model["n_eval_episodes"], model["qtable"], model["eval_seed"])

    readme_path = repo_local_path / "README.md"
    readme = ""
    print(readme_path.exists())
    if readme_path.exists():
        with readme_path.open("r", encoding="utf8") as f:
            readme = f.read()
    else:
        readme = model_card

    with readme_path.open("w", encoding="utf-8") as f:
        f.write(readme)

    # Sauvegarder les m√©triques dans les m√©tadonn√©es du Readme
    metadata_save(readme_path, metadata)

    # √âtape 6 : Enregistrer une vid√©o
    video_path = repo_local_path / "replay.mp4"
    record_video(env, model["qtable"], video_path, video_fps)

    # √âtape 7 : Envoyer le tout sur le Hub
    api.upload_folder(
        repo_id=repo_id,
        folder_path=repo_local_path,
        path_in_repo=".",
    )

    print("Your model is pushed to the Hub. You can view your model here: ", repo_url)
````

Once the passenger is dropped off, the episode ends.

# Part 3: Load from Hub üîΩ

What's amazing with Hugging Face Hub ü§ó is that you can easily load powerful models from the community.

Loading a saved model from the Hub is really easy:

1. You go https://huggingface.co/models?other=q-learning to see the list of all the q-learning saved models.
2. You select one and copy its repo_id

<img src="https://huggingface.co/datasets/huggingface-deep-rl-course/course-images/resolve/main/en/notebooks/unit2/copy-id.png" alt="Copy id">

3. Then we just need to use `load_from_hub` with:
- The repo_id
- The filename: the saved model inside the repo.

#### Do not modify this code

````python
# Fonction pour charger un mod√®le depuis le Hub
from urllib.error import HTTPError

from huggingface_hub import hf_hub_download


def load_from_hub(repo_id: str, filename: str) -> str:
    """
    T√©l√©charge un mod√®le depuis le Hugging Face Hub.
    :param repo_id: id du d√©p√¥t
    :param filename: nom du fichier mod√®le
    """
    # R√©cup√©rer le mod√®le du Hub
    pickle_model = hf_hub_download(
        repo_id=repo_id,
        filename=filename
    )

    with open(pickle_model, 'rb') as f:
      downloaded_model_file = pickle.load(f)

    return downloaded_model_file
````
